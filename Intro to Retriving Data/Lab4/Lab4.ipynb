<h2 style="color:red;"><b>Naive Bayes Classifier</b></h2>
<p>A Naive Bayes classifier is a probabilistic machine learning model that’s used for classification task.Below you can see a formula for the calculation:</p>
$$
    P(A|B)=\frac{P(B|A)*P(A)}{P(B)}
$$
<p>Using Bayes theorem, we can find the probability of A happening, given that B has occurred. </p>
<p>In various applications such as spam filtering, text classification, sentiment analysis, and recommendation systems, Naive Bayes classifier is used successfully. It reccomend itself as fastest classification algorithm for a large chunk of data. </p>
<ol>We There are 3 types of Naïve Bayes algorithm:
    <li>Gaussian Naïve Bayes:use when dealing with continuous data.</li>
    <li>Multinomial Naïve Bayes:mostly used for document classification problem(whether a document belongs to the some category ). The features/predictors used by the classifier are the frequency of the words present in the document.</li>
    <li>Bernoulli Naïve Bayes:similar to the multinomial naive bayes but the predictors are boolean variables.</li>
</ol>
<ol>Advantages:
    <li>Works quickly.</li>
    <li>Suitable for solving multi-class prediction problems. </li>
    <li>If its assumption of the independence of features holds true, it can perform better than other models and requires much less training data. </li>
    <li>Naive Bayes is better suited for categorical input variables than numerical variables.</li>
</ol>
<ol>Disadvantages:
    <li>Asumes that all predictors (or features) are independent, rarely happening in real life.</li>
    <li>Suitable for solving multi-class prediction problems. </li>
    <li>Faces the ‘zero-frequency problem’ </li>
    <li>Estimations can be wrong in some cases.</li>
</ol>

<h2 style="color:red;"><b>Feature Selection</b></h2>
<ul>
    <li>Definition:It's the process of isolating the most consistent, non-redundant, and relevant features to use in model construction.</li>
    <li>In text classificationit can improve the scalability, efficiency and accuracy of a text classifier.</li>
    <li>For feature selection we can use- sklearn.feature_selection module.</li>
    <li>Filter Methods-rely on the features’ characteristics without using any machine learning algorithm,always used at the beginning of the selection process to get rid of the irrelevant, duplicated, correlated, and constant features:
        <ol>
            <li>Advantages:selected features can be used in any machine learning algorithm and you can process thousands of features in a matter of seconds.</li>
            <li>The mehtods:Basic Filter Methods,Correlation Filter Methods,Statistical and Ranking Filter Methods.</li>
        </ol>
    </li>
    <li>Wrapper methods-usually result in better predictive accuracy than filter methods.: <ol>
            <li>Advantages:detect the interaction between variables and find the optimal feature subset for the desired machine learning algorithm</li>
            <li>The process:Search for a subset of features->Build a machine learning model->Evaluate model performance->Repeat
        </li>
        <li>Stopping Criteria:need to be defined by the developer.</li>
        </ol>
    </li>
    <li>Embedded methods:<ol>
            <li>Advantages:take into consideration the interaction of features like wrapper methods do,faster,more accurate than filter methods.</li>
            <li>The process:train a model->derive feature importance from this model->remove non-important features using the derived feature importance.</li>
        </ol>
    </li>
    <li>We need use feature selection to text classification because features consume a lot of time and computing power to get the result. Particular in case of text where features are words.Using it helps a lot to Naive Bayes classifier because he is exspensive to train.</li>
    <li>Feature selection is necessary for multivariate Bernoulli NB for avoiding noise and multi-counting.
</li>
</ul>
