```python
import pandas as pd
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
spec_char_list=['~', ':',  '+',  '@', '^', '%','*', '|', '&','<', '`', '=', '!', '>','?', '#', '$']
#-------------------------------------------------------------------------------------------------------------------------------
file=pd.read_csv('trainTwitter.csv')
tweetCol=file['tweet']
tweetTable=list()
#-------------------------------------------------------------------------------------------------------------------------------
def wordsNumber(str):
    temp=str.split()
    for s in temp:
        if s[0]=='@' or s[0]=='#':
            temp.remove(s)
    return len(temp)
#-------------------------------------------------------------------------------------------------------------------------------
def lettersNumber(str):
    temp=str.split()
    num=0
    for s in temp:
        if s[0]=='@' or s[0]=='#':
            temp.remove(s)
    for t in temp:
            num=num+len(t)
    return num
#-------------------------------------------------------------------------------------------------------------------------------
def stopWordsCount(str):
    stop_words = set(stopwords.words('english'))
    word_tokens = str.split()
    num=0
    for s in word_tokens:
        if s[0]=='@' or s[0]=='#':
            word_tokens.remove(s)
    for s in word_tokens:
        if s in stop_words:
            num=num+1
    return num
#-------------------------------------------------------------------------------------------------------------------------------
def numCharCount(str):
    word_tokens = str.split()
    total_num_char=0
    for s in word_tokens:
        if s[0]=='@' or s[0]=='#':
            word_tokens.remove(s)
            
    for s in word_tokens:
        for i in s:
            if i.isnumeric():
                total_num_char=total_num_char+1
    return total_num_char
#-------------------------------------------------------------------------------------------------------------------------------
def specCharCount(str):
    word_tokens = str.split()
    total_spec_char=0
    for s in word_tokens:
        if s[0]=='@' or s[0]=='#':
            word_tokens.remove(s)
    for s in word_tokens:
        for i in s:
            if i in spec_char_list:
                total_spec_char+=total_spec_char+1
    return total_spec_char
#-------------------------------------------------------------------------------------------------------------------------------
def capWordsCount(str):
    word_tokens = str.split()
    total_cap_words=0
    for s in word_tokens:
        if s[0]=='@' or s[0]=='#':
            word_tokens.remove(s)
    for s in word_tokens:
        if s.isupper():
            total_cap_words=total_cap_words+1
    return total_cap_words
#-------------------------------------------------------------------------------------------------------------------------------
def makingRow(curTweet):
    result=list()
    result.append(curTweet)
    #WordsNumber append
    result.append(wordsNumber(curTweet))
    #LettersNumber append
    result.append(lettersNumber(curTweet))
    #Avg_Word_Size append
    if wordsNumber(curTweet) == 0:
        result.append(0)
    else:
        result.append(round(lettersNumber(curTweet)/wordsNumber(curTweet),2))
    #Number of stopwords append
    result.append(stopWordsCount(curTweet))
    #Number of numeric characters append
    result.append(numCharCount(curTweet))
    #Number of special characters append
    result.append(specCharCount(curTweet))
    #Number of capital words append
    result.append(capWordsCount(curTweet))
    return result
#-------------------------------------------------------------------------------------------------------------------------------
def mainFunc():
    i=0
    while(i<len(tweetCol)):
        current=makingRow(tweetCol[i])
        tweetTable.append(current)
        i=i+1
#-------------------------------------------------------------------------------------------------------------------------------

mainFunc()
#create table with inf for each tweet
df = pd.DataFrame(tweetTable, columns =['Tweet','WordsNumber', 'LettersNumber','Avg_Word_Size','Number of stopwords',
                                    'Number of numeric characters','Number of special characters','Number of capital words'])
df.head()
```
